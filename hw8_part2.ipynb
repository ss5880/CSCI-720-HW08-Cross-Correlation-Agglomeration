```python
"""
@Author: Sumire Sakai, Tejal Shanbhag
@Purpose: To perform the Agglomeration of the data
"""

import pandas as pd
import numpy as np

def read_Data_file():
    data = pd.read_csv('HW_PCA_SHOPPING_CART_v896.csv')
    columns=list(data.columns[1:])
    rows=list(data.iloc[:, 0])
    return data,rows,columns

def getEuclideanDistance(x,y):
    distance = np.sqrt(np.sum([(a - b) * (a - b) for a, b in zip(x, y)]))
    return distance

def generateDistanceMatrix(data,rows,coloumns):
    DISTANCE_MATRIX=[[float('inf') for guest_x in range(len(rows))]for guest_y in range(len(rows))]
    for guest_index_x in range(0, len(rows)):
        for guest_index_y in range(0, len(rows)):
            if guest_index_x>guest_index_y:
                row1=data.loc[guest_index_x]
                row2=data.loc[guest_index_y]
                d=getEuclideanDistance(row1,row2)
                DISTANCE_MATRIX[guest_index_x][guest_index_y]=d
                DISTANCE_MATRIX[guest_index_y][guest_index_x] = d
    headers = [i for i in rows]
    df = pd.DataFrame(DISTANCE_MATRIX, columns=headers, index=headers)
    df.to_csv('Distance_Matrix.csv')
    return df

    


```

```python
data,rows,cols=read_Data_file()
```

```python
GUEST_DISTANCE_MATRIX=generateDistanceMatrix(data,rows,cols)
GUEST_DISTANCE_MATRIX
```
